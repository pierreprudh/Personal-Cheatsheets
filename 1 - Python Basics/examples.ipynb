### String Manipulation

```python
name = "John"
greeting = f"Hello, {name}!"   # f-string
length = len(name)

print(greeting, length)
```

```python
"hello".upper()   # "HELLO"
"Hello".lower()   # "hello"
"data".capitalize()   # "Data"
```
