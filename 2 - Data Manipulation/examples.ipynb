Here we look at the target correlation : 
- `petal length (cm)`,	`petal width (cm)` are extremely correlated 

What does this mean ? 

- As petal length and width increase, the class label increases (class changement)
- Petal width is the single strongest linear indicator of class
- Sepal width decreases as we move to higher-class labels (negativ mid correlation)

**Petal features alone could classify very well**

But we have multicollinearity also (very strong correlation between petal and sepal information).
So we might need to reduce the features numbers.

### Outliers in petal lenght 

```python
q1 = df["petal length (cm)"].quantile(0.25)
q3 = df["petal length (cm)"].quantile(0.75)
iqr = q3 - q1

outliers = df[(df["petal length (cm)"] < q1 - 1.5 * iqr) |
              (df["petal length (cm)"] > q3 + 1.5 * iqr)]

display(outliers)
```

In this easy dataset we don't face this type of problem !
